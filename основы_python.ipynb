#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s):
    vowels = set('aeiou')
    return sum(1 for char in s.lower() if char in vowels)

assert count_vowels("hello") == 2
assert count_vowels("AEIOU") == 5
assert count_vowels("Python") == 1
assert count_vowels("bcdfg") == 0
assert count_vowels("") == 0
assert count_vowels("AaEeIiOoUu") == 10
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def all_unique(s):
    return len(s) == len(set(s))

assert all_unique("") == True
assert all_unique("abc") == True
assert all_unique("abca") == False
assert all_unique("12345") == True
assert all_unique("11223") == False
assert all_unique("aA") == True
assert all_unique("!@#") == True
assert all_unique("!!") == False
```

# Задача 3

бит равных 1 в этом числе. Написать тесты для кода

```python
def count_set_bits(n):
    return bin(n).count('1')

assert count_set_bits(1) == 1
assert count_set_bits(2) == 1
assert count_set_bits(3) == 2
assert count_set_bits(4) == 1
assert count_set_bits(5) == 2
assert count_set_bits(7) == 3
assert count_set_bits(8) == 1
assert count_set_bits(15) == 4
assert count_set_bits(16) == 1
assert count_set_bits(0) == 0
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    steps = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        steps += 1
    return steps

assert magic(39) == 3
assert magic(4) == 0
assert magic(999) == 4
assert magic(10) == 1
assert magic(25) == 2
assert magic(0) == 0
assert magic(99) == 2
assert magic(777) == 4
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    n = len(pred)
    return sum((p - t) ** 2 for p, t in zip(pred, true)) / n

assert mse([0, 0, 0], [0, 0, 0]) == 0.0
assert mse([1, 2, 3], [1, 2, 3]) == 0.0
assert mse([1, 1, 1], [0, 0, 0]) == 1.0
assert mse([2, 2, 2], [0, 0, 0]) == 4.0
assert mse([1, 2, 3], [2, 3, 4]) == 1.0
assert mse([1.5, 2.0], [1.0, 1.5]) == 0.25
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factorization_string(n):
    result = ""
    factor = 2
    while factor * factor <= n:
        count = 0
        while n % factor == 0:
            n //= factor
            count += 1
        if count > 0:
            if count == 1:
                result += f"({factor})"
            else:
                result += f"({factor}**{count})"
        factor += 1
    if n > 1:
        result += f"({n})"
    return result

assert prime_factorization_string(86240) == "(2**5)(5)(7**2)(11)"
assert prime_factorization_string(2) == "(2)"
assert prime_factorization_string(3) == "(3)"
assert prime_factorization_string(4) == "(2**2)"
assert prime_factorization_string(6) == "(2)(3)"
assert prime_factorization_string(8) == "(2**3)"
assert prime_factorization_string(9) == "(3**2)"
assert prime_factorization_string(12) == "(2**2)(3)"
assert prime_factorization_string(1) == ""
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    total = 0
    k = 0
    while total < number:
        k += 1
        total += k * k
    if total == number and number != 0:
        return k
    return "It is impossible"

assert pyramid(1) == 1
assert pyramid(5) == 2
assert pyramid(14) == 3
assert pyramid(30) == 4
assert pyramid(55) == 5
assert pyramid(2) == "It is impossible"
assert pyramid(3) == "It is impossible"
assert pyramid(4) == "It is impossible"
assert pyramid(0) == "It is impossible"
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def balanced_num(n):
    s = str(n)
    length = len(s)
    if length % 2 == 0:
        mid = length // 2
        left = s[:mid-1]
        right = s[mid+1:]
    else:
        mid = length // 2
        left = s[:mid]
        right = s[mid+1:]
    sum_left = sum(int(d) for d in left) if left else 0
    sum_right = sum(int(d) for d in right) if right else 0
    return sum_left == sum_right

assert balanced_num(12321) == True
assert balanced_num(1230) == False
assert balanced_num(1221) == True
assert balanced_num(1111) == True
assert balanced_num(12345) == False
assert balanced_num(1) == True
assert balanced_num(12) == True
assert balanced_num(123) == False
assert balanced_num(123321) == True
assert balanced_num(112233) == False
```
